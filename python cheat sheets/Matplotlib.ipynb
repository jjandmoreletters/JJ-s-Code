## two options for subplots

```python
#subplot option 1
fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))

#plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_better_prices.keys(), nut_better_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax=plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))

#plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_better_prices.keys(), nut_better_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which one should I use (pyplot vs matplotlib 00 method?)

* When plotting quick use pyplot method
* When plotting something more advanced use 00 method

```python
over_50=heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
#Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
##00 method
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x='age',
             y='chol',
             c='target',
             ax=ax);

#ax.set_xlim([45,100]);
```

```python
#00 from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);

#customize
ax.set(title="Heart disease and Cholesterol Levels",
       xlabel="age",
       ylabel="Cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="target");

#add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="-");
```

```python
# subplots of chol, age, thalach

#00 from scratch
fig, (ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,6),
                           sharex=True)

#plot the data ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"]);

scatter2=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"]);
#customize ax0
ax0.set(title="Heart disease and Cholesterol Levels",
       #xlabel="age",
       ylabel="Cholesterol");

ax1.set(title="Heart disease and thalach",
       xlabel="age",
       ylabel="thalach");
#add a legend
ax0.legend(*scatter.legend_elements(), title="target");

ax1.legend(*scatter2.legend_elements(), title="target");

#add horizontal line
ax0.axhline(over_50["chol"].mean(),
           linestyle="-");

ax1.axhline(over_50["thalach"].mean(),
           linestyle="--");

fig.suptitle("Heart Disease", fontsize=16, fontweight="bold");
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

#00 from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                  cmap="winter");

#customize
ax.set(title="Heart disease and Cholesterol Levels",
       xlabel="age",
       ylabel="Cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="target");

#add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="-");
```

```python
# subplots of chol, age, thalach

#00 from scratch
fig, (ax0,ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,6),
                           sharex=True)

#plot the data ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter");

scatter2=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                    cmap="summer");
#customize ax0
ax0.set(title="Heart disease and Cholesterol Levels",
       #xlabel="age",
       ylabel="Cholesterol");
ax0.set_xlim([50,80])
ax0.set_ylim([60,600])
ax1.set(title="Heart disease and thalach",
       xlabel="age",
       ylabel="thalach");
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#add a legend
ax0.legend(*scatter.legend_elements(), title="target");

ax1.legend(*scatter2.legend_elements(), title="target");

#add horizontal line
ax0.axhline(over_50["chol"].mean(),
           linestyle="-");

ax1.axhline(over_50["thalach"].mean(),
           linestyle="--");

fig.suptitle("Heart Disease", fontsize=16, fontweight="bold");
```
